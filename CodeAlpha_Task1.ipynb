```python
# Data Redundancy Removal System

class DataRedundancySystem:

    # Constructor
    def __init__(self):

        # Simulated cloud database
        self.database = []

    # Validate Data
    def validate_data(self, data):

        # Data should not be empty
        if not data:
            return False

        # Required fields
        required_fields = ["id", "name", "email"]

        # Check required fields
        for field in required_fields:

            if field not in data:
                return False

            if data[field] == "":
                return False

        return True

    # Check Duplicate Data
    def is_duplicate(self, new_data):

        for record in self.database:

            # Compare ID or Email
            if (record["id"] == new_data["id"] or
                    record["email"] == new_data["email"]):

                return True

        return False

    # Add Data
    def add_data(self, data):

        # Step 1: Validation
        if not self.validate_data(data):

            print("Invalid Data Found:", data)
            return

        # Step 2: Duplicate Check
        if self.is_duplicate(data):

            print("Redundant Data Found:", data)
            return

        # Step 3: Store Unique Data
        self.database.append(data)

        print("Unique Data Added Successfully:", data)

    # Display Database
    def show_database(self):

        print("\nCurrent Database Records:\n")

        for record in self.database:
            print(record)


# ---------------- MAIN PROGRAM ----------------

system = DataRedundancySystem()

# Sample Data
data1 = {
    "id": 1,
    "name": "Rahul",
    "email": "rahul@gmail.com"
}

data2 = {
    "id": 2,
    "name": "Aman",
    "email": "aman@gmail.com"
}

# Duplicate Data
data3 = {
    "id": 1,
    "name": "Rahul",
    "email": "rahul@gmail.com"
}

# Invalid Data
data4 = {
    "id": 3,
    "name": "",
    "email": "test@gmail.com"
}

# Add Data
system.add_data(data1)
system.add_data(data2)
system.add_data(data3)
system.add_data(data4)

# Show Final Database
system.show_database()
```
